```python
#Adaptive Food Suggestion Engine by Fuzzy Logic

import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl 

taste = ctrl.Antecedent(np.arange(0,11,0.5),'taste') # thuc an theo thang diem 0-10
location = ctrl.Antecedent(np.arange(0,3.2,0.1), 'location') # khoang cach tu 0 den 3 km
budget = ctrl.Antecedent(np.arange(0,9.2,0.5), 'budget') # ngan sach tu 0-9$
preference_level = ctrl.C=ctrl.Consequent(np.arange(0,11,0.5), 'preference_level') # cap do tu 0-10 

taste['low'] = fuzz.trimf(taste.universe, [0,2,4]) 
taste['medium'] = fuzz.trimf(taste.universe, [3,5,7]) 
taste['high'] = fuzz.trimf(taste.universe, [6,7.5,10]) 
#taste.view()

location['near'] = fuzz.trimf(location.universe, [0,0.5,1])
location['medium'] = fuzz.trimf(location.universe, [0.8,1.5,2]) 
location['far'] = fuzz.trimf(location.universe, [1.8,3,3]) 
#location.view()

budget['low'] = fuzz.trimf(budget.universe, [0,1.5,3])
budget['medium'] = fuzz.trimf(budget.universe, [3,4.5,6]) 
budget['high'] = fuzz.trimf(budget.universe, [6,7.5,9]) 
#budget.view()

preference_level['less'] = fuzz.trimf(preference_level.universe, [0,1.5,3]) 
preference_level['medium'] = fuzz.trimf(preference_level.universe, [2,5,7])
preference_level['high'] = fuzz.trimf(preference_level.universe, [6,8,10]) 
#preference_level.view()

rule1 = ctrl.Rule(taste['low'] & location['near'] & budget['low'], preference_level['medium'])
rule2 = ctrl.Rule(taste['low'] & location['near'] & budget['medium'], preference_level['less'])
rule3 = ctrl.Rule(taste['low'] & location['near'] & budget['high'], preference_level['less'])
rule4 = ctrl.Rule(taste['low'] & location['medium'] & budget['low'], preference_level['medium'])
rule5 = ctrl.Rule(taste['low'] & location['medium'] & budget['medium'], preference_level['less'])
rule6 = ctrl.Rule(taste['low'] & location['medium'] & budget['high'], preference_level['less'])
rule7 = ctrl.Rule(taste['low'] & location['far'] & budget['low'], preference_level['less'])
rule8 = ctrl.Rule(taste['low'] & location['far'] & budget['medium'], preference_level['less'])
rule9 = ctrl.Rule(taste['low'] & location['far'] & budget['high'], preference_level['less'])
rule10 = ctrl.Rule(taste['medium'] & location['near'] & budget['low'], preference_level['high'])
rule11 = ctrl.Rule(taste['medium'] & location['near'] & budget['medium'], preference_level['medium'])
rule12 = ctrl.Rule(taste['medium'] & location['near'] & budget['high'], preference_level['medium'])
rule13 = ctrl.Rule(taste['medium'] & location['medium'] & budget['low'], preference_level['high'])
rule14 = ctrl.Rule(taste['medium'] & location['medium'] & budget['medium'], preference_level['medium'])
rule15 = ctrl.Rule(taste['medium'] & location['medium'] & budget['high'], preference_level['medium'])
rule16 = ctrl.Rule(taste['medium'] & location['far'] & budget['low'], preference_level['medium'])
rule17 = ctrl.Rule(taste['medium'] & location['far'] & budget['medium'], preference_level['medium'])
rule18 = ctrl.Rule(taste['medium'] & location['far'] & budget['high'], preference_level['less'])
rule19 = ctrl.Rule(taste['high'] & location['near'] & budget['low'], preference_level['high'])
rule20 = ctrl.Rule(taste['high'] & location['near'] & budget['medium'], preference_level['high'])
rule21 = ctrl.Rule(taste['high'] & location['near'] & budget['high'], preference_level['medium'])
rule22 = ctrl.Rule(taste['high'] & location['medium'] & budget['low'], preference_level['high'])
rule23 = ctrl.Rule(taste['high'] & location['medium'] & budget['medium'], preference_level['high'])
rule24 = ctrl.Rule(taste['high'] & location['medium'] & budget['high'], preference_level['medium'])
rule25 = ctrl.Rule(taste['high'] & location['far'] & budget['low'], preference_level['high'])
rule26 = ctrl.Rule(taste['high'] & location['far'] & budget['medium'], preference_level['medium'])
rule27 = ctrl.Rule(taste['high'] & location['far'] & budget['high'], preference_level['less'])

preference_leveling_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,
                                   rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,
                                   rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25,rule26,rule27])
preference_leveling =ctrl.ControlSystemSimulation(preference_leveling_ctrl)
preference_leveling.input['taste']=2.5
preference_leveling.input['location']=2
preference_leveling.input['budget']=0.5
preference_leveling.compute()
print(preference_leveling.output['preference_level'])
preference_level.view(sim=preference_leveling)
```

```python
# COOKER

import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl 

time = ctrl.Antecedent(np.arange(10,181,1),'time') # thoi gian nau 0-180p
rice = ctrl.Antecedent(np.arange(0.2,1.1,0.1), 'rice') # khoi luong gao 0-1kg
power = ctrl.C=ctrl.Consequent(np.arange(0,102,1), 'power') # cong suat nau 0-100%

time['VF'] = fuzz.trimf(time.universe, [10,12,15]) # thoi gian rat nhanh
time['F'] = fuzz.trimf(time.universe, [12,15,30]) # thoi gian nhanh
time['AV'] = fuzz.trimf(time.universe, [15,30,45]) #thoi gian trung binh
time['S'] = fuzz.trimf(time.universe, [30,45,120]) #thoi gian lau
time['VS'] = fuzz.trimf(time.universe, [45,120,180]) #thoi gian rat lau
#time.view()

rice['VL'] = fuzz.trimf(rice.universe, [0.2,0.3,0.4]) # gao rat it
rice['L'] = fuzz.trimf(rice.universe, [0.3,0.4,0.6]) #gao it
rice['AV'] = fuzz.trimf(rice.universe, [0.4,0.6,0.8]) #gao trung binh
rice['M'] = fuzz.trimf(rice.universe, [0.6,0.8,0.9]) #gao nhieu
rice['VM'] = fuzz.trimf(rice.universe, [0.8,0.9,1]) #gao rat nhieu
#rice.view()

power['VS'] = fuzz.trimf(power.universe, [0,0,20]) # cong suat rat nho
power['S'] = fuzz.trimf(power.universe, [10,20,40]) # cong suat nho
power['AV'] = fuzz.trimf(power.universe, [20,40,60]) #cong suat trung binh
power['H'] = fuzz.trimf(power.universe, [40,60,80]) #cong suat lon
power['VH'] = fuzz.trimf(power.universe, [60,100,100]) #cong suat rat lon
#power.view()

rule1 = ctrl.Rule(time['VF'] & rice['VL'], power['AV'])
rule2 =ctrl.Rule(time['VF'] & rice['L'], power['H'])
rule3 = ctrl.Rule(time['VF'] & rice['AV'], power['VH'])
rule4 = ctrl.Rule(time['VF'] & rice['M'], power['VH'])
rule5 = ctrl.Rule(time['VF'] & rice['VM'], power['VH'])
rule6 = ctrl.Rule(time['F'] & rice['VL'], power['AV'])
rule7 = ctrl.Rule(time['F'] & rice['L'], power['AV'])
rule8 = ctrl.Rule(time['F'] & rice['AV'], power['H'])
rule9 = ctrl.Rule(time['F'] & rice['M'], power['VH'])
rule10 = ctrl.Rule(time['F'] & rice['VM'], power['VH'])
rule11 = ctrl.Rule(time['AV'] & rice['VL'], power['AV'])
rule12 = ctrl.Rule(time['AV'] & rice['L'], power['AV'])
rule13 = ctrl.Rule(time['AV'] & rice['AV'], power['AV'])
rule14 = ctrl.Rule(time['AV'] & rice['M'], power['H'])
rule15 = ctrl.Rule(time['AV'] & rice['VM'], power['H'])
rule16 = ctrl.Rule(time['S'] & rice['VL'], power['S'])
rule17 = ctrl.Rule(time['S'] & rice['L'], power['S'])
rule18 = ctrl.Rule(time['S'] & rice['AV'], power['S'])
rule19 = ctrl.Rule(time['S'] & rice['M'], power['AV'])
rule20 = ctrl.Rule(time['S'] & rice['VM'], power['AV'])
rule21 = ctrl.Rule(time['VS'] & rice['VL'], power['VS'])
rule22 = ctrl.Rule(time['VS'] & rice['L'], power['VS'])
rule23 = ctrl.Rule(time['VS'] & rice['AV'], power['VS'])
rule24 = ctrl.Rule(time['VS'] & rice['M'], power['S'])
rule25 = ctrl.Rule(time['VS'] & rice['VM'], power['S'])

powering_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,
                                   rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,
                                   rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
powering =ctrl.ControlSystemSimulation(powering_ctrl)
powering.input['time']=35
powering.input['rice']=0.4
powering.compute()
print(powering.output['power'])
power.view(sim=powering)
```

```python
# IMPROTA
#case study1

import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl 

glycaemia = ctrl.Antecedent(np.arange(50,151,1),'glycaemia') # glycaemia 50-150 mg/dl
sirolimus = ctrl.Antecedent(np.arange(0,21,0.5), 'sirolimus') # sirolimus 0-2000 ng/ml
proteinuria = ctrl.C=ctrl.Consequent(np.arange(50,3600,1), 'proteinuria') # proteinuria 100-3500mg/24h

glycaemia['good'] = fuzz.trapmf(glycaemia.universe, [20,20,70,99]) 
glycaemia['alarm'] = fuzz.trimf(glycaemia.universe, [60,100,140]) 
glycaemia['danger'] = fuzz.trapmf(glycaemia.universe, [125,126,150,150]) 
#glycaemia.view()

sirolimus['alarm-down'] = fuzz.trimf(sirolimus.universe, [0,2.5,5]) 
sirolimus['suff'] = fuzz.trimf(sirolimus.universe, [4,5.5,7]) 
sirolimus['good'] = fuzz.trimf(sirolimus.universe, [6,9,12]) 
sirolimus['alarm-up'] = fuzz.trimf(sirolimus.universe, [11,13,14]) 
sirolimus['danger'] = fuzz.trimf(sirolimus.universe, [13,16,20]) 
#sirolimus.view()

proteinuria['good'] = fuzz.trapmf(proteinuria.universe, [50,50,150,150]) 
proteinuria['suff'] = fuzz.trimf(proteinuria.universe, [150,290,500]) 
proteinuria['alterate'] = fuzz.trimf(proteinuria.universe, [500,750,1000]) 
proteinuria['alarm'] = fuzz.trimf(proteinuria.universe, [1000,2000,3000]) 
proteinuria['danger'] = fuzz.trapmf(proteinuria.universe, [2800,3000,3500,3500])
#proteinuria.view()

rule1 = ctrl.Rule( glycaemia['good'] & sirolimus['alarm-down'], proteinuria['suff'])
rule2 =ctrl.Rule( glycaemia['good'] & sirolimus['suff'], proteinuria['suff'])
rule3 = ctrl.Rule( glycaemia['good'] & sirolimus['good'], proteinuria['good'])
rule4 = ctrl.Rule( glycaemia['good'] & sirolimus['alarm-up'], proteinuria['alterate'])
rule5 = ctrl.Rule( glycaemia['good'] & sirolimus['danger'], proteinuria['danger'])
rule6 = ctrl.Rule( glycaemia['alarm'] & sirolimus['alarm-down'], proteinuria['suff'])
rule7 = ctrl.Rule( glycaemia['alarm'] & sirolimus['suff'], proteinuria['alterate'])
rule8 = ctrl.Rule( glycaemia['alarm'] & sirolimus['good'], proteinuria['suff'])
rule9 = ctrl.Rule( glycaemia['alarm'] & sirolimus['alarm-up'], proteinuria['alarm'])
rule10 = ctrl.Rule( glycaemia['alarm'] & sirolimus['danger'], proteinuria['danger'])
rule11 = ctrl.Rule( glycaemia['danger'] & sirolimus['alarm-down'], proteinuria['suff'])
rule12 = ctrl.Rule( glycaemia['danger'] & sirolimus['suff'], proteinuria['danger'])
rule13 = ctrl.Rule( glycaemia['danger'] & sirolimus['good'], proteinuria['suff'])
rule14 = ctrl.Rule( glycaemia['danger'] & sirolimus['alarm-up'], proteinuria['danger'])
rule15 = ctrl.Rule( glycaemia['danger'] & sirolimus['danger'], proteinuria['danger'])

proteinuriaing_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,
                                   rule10,rule11,rule12,rule13,rule14,rule15])
proteinuriaing =ctrl.ControlSystemSimulation(proteinuriaing_ctrl)
proteinuriaing.input['glycaemia']=100
proteinuriaing.input['sirolimus']=15
proteinuriaing.compute()
print(proteinuriaing.output['proteinuria'])
proteinuria.view(sim=proteinuriaing)
```
