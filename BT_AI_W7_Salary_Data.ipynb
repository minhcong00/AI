```python
import pandas as pd
url = '/content/Salary_Data.csv'
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
sns.pairplot(data)
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation,annot=True)
```

```python
# Hoi quy bac 1
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Salary']),y=data['YearsExperience'])
print('Gia tri,Intercept la:',model.intercept_)
print('Gia tri coeffient la:',model.coef_)
model.score(pd.DataFrame(data['Salary']),data['YearsExperience'])

data.plot(kind='scatter',x='Salary',y='YearsExperience',figsize=(9,9),color='black')
y_pred=model.predict(X=pd.DataFrame(data['Salary']))
plt.plot(data['Salary'],y_pred,color='green')
```

```python
#hoi quy bac 2
import numpy as np
model=linear_model.LinearRegression()
X= pd.DataFrame([data['Salary'],data['Salary']**2]).T
model.fit(X,y=data['YearsExperience'])
print('He so Intercept la:',model.intercept_)
print('He so coeffient la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(pd.DataFrame([data['Salary'],data['Salary']**2]).T,data['YearsExperience']))

range = np.arange(data['Salary'].min(),data['Salary'].max(),0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize = (9,9), color = 'blue', xlim = 0.7)
plt.plot(range, y_pred, color = 'red')
```

```python
#Hồi quy bậc 5
model = linear_model.LinearRegression()
X=pd.DataFrame([data['Salary'], data['Salary']**2, data['Salary']**3, data['Salary']**4, data['Salary']**5]).T
y=data['YearsExperience']
model.fit(X, y)
print('Intercept của mô hình : ', model.intercept_)
print('Coefient của mô hình : ', model.coef_)
print('Độ chính xác của mô hình : ', model.score(X,y))

range = np.arange(data['Salary'].min(), data['Salary'].max(), 0.1)
data.plot(kind='scatter', x='Salary', y='YearsExperience', figsize=(9,9), color='black', xlim=0.7)
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_pred = model.predict(X=predictor)
plt.plot(range, y_pred, color='blue')
```
