```python
import pandas as pd
url='http://winterolympicsmedals.com/medals.csv'
data= pd.read_csv(url)
data
```

```python
data=data.set_index(data['NOC'])
data.loc['USA']
```

```python
# tổng số huy chương của mỹ từ năm 1924-2006
USA = data[(data['NOC']== 'USA')&(data['Medal'])]
print(' So huy chuong cua My tu nam 1924 den nam 2006 la:', USA['Medal'].count())
```

```python
# SỐ Huy chương vàng bạc đồng của mỹ tư năm 1924-2006 của nam và nữ
USA = data[(data['NOC']== 'USA')&(data['Medal']=='Bronze')&(data['Year']>= 1924) & (data['Year']<=2006) & (data['Event gender']=='M')]
print(' So huy chuong dong cua My tu nam 1924 den nam 2006 cua nam la:', USA['Medal'].count())
USA = data[(data['NOC']== 'USA')&(data['Medal']=='Silver')&(data['Year']>= 1924) & (data['Year']<=2006) & (data['Event gender']=='M')]
print(' So huy chuong bac cua My tu nam 1924 den nam 2006 cua nam la:', USA['Medal'].count())
USA = data[(data['NOC']== 'USA')&(data['Medal']=='Gold')&(data['Year']>= 1924) & (data['Year']<=2006) & (data['Event gender']=='M')]
print(' So huy chuong vang cua My tu nam 1924 den nam 2006 cua nam la:', USA['Medal'].count())
USA = data[(data['NOC']== 'USA')&(data['Medal']=='Bronze')&(data['Year']>= 1924) & (data['Year']<=2006) & (data['Event gender']=='W')]
print(' So huy chuong dong cua My tu nam 1924 den nam 2006 cua nu la:', USA['Medal'].count())
USA = data[(data['NOC']== 'USA')&(data['Medal']=='Silver')&(data['Year']>= 1924) & (data['Year']<=2006) & (data['Event gender']=='W')]
print(' So huy chuong bac cua My tu nam 1924 den nam 2006 cua nu la:', USA['Medal'].count())
USA = data[(data['NOC']== 'USA')&(data['Medal']=='Gold')&(data['Year']>= 1924) & (data['Year']<=2006) & (data['Event gender']=='W')]
print(' So huy chuong vang cua My tu nam 1924 den nam 2006 cua nu la:', USA['Medal'].count())
```
